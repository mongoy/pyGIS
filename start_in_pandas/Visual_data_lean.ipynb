## Визуализация Pandas
Датасет- https://www.kaggle.com/open-powerlifting/powerlifting-database

```python
# читаем датасет
df_opl = pd.read_csv(os.path.join(ROOT_PATH,'openpowerlifting.csv'), sep=',')
```

```python
df_opl.head()
```

```python
df_opl.info()
```

```python
df_opl.columns
```

```python
# выберем ненулевые колонки
cols = ['Equipment', 'Age', 'BodyweightKg', 'WeightClassKg', 'Best3SquatKg', 'Best3BenchKg',
       'Best3DeadliftKg', 'Place', 'TotalKg', 'Sex']
df_opl = df_opl[cols]
df_opl
```

### одномерные данные

```python
# BAR Chat - столбчатая диаграмма
fig = plt.figure(figsize=(14,5))
df_opl['Equipment'].value_counts().plot.bar()
```

```python
# BAR Chat - столбчатая диаграмма + сортировка по индексу (по алфавиту)
fig = plt.figure(figsize=(14,5))
df_opl['Equipment'].value_counts().sort_index().plot.bar()
```

```python
# BAR Chat - столбчатая диаграмма + относительные частоты распределения (делим величину на длинну всей выборки)
fig = plt.figure(figsize=(14,5))
(df_opl['Equipment'].value_counts() / len(df_opl)).plot.bar()
```

```python
# BAR Chat - столбчатая диаграмма + группировка по 'Equipment' -> средние значения
df_opl.groupby(['Equipment'])[['BodyweightKg', 'TotalKg']].mean()
```

```python
# BAR Chat - столбчатая диаграмма stacked=False (наслоение)
df_opl.groupby(['Equipment'])[['BodyweightKg', 'TotalKg']].mean().plot.bar(figsize=(14,5), stacked=False)
```

```python
# ИЛИ stacked=True (наслоение)
df_opl.groupby(['Equipment'])[['BodyweightKg', 'TotalKg']].mean().plot.bar(figsize=(14,5), stacked=True)
```

```python
# линейный график (более 20 категориальных значений)
fig = plt.figure(figsize=(14,5))
(df_opl['Equipment'].value_counts() / len(df_opl)).plot.line()
```

```python
# Площадь
fig = plt.figure(figsize=(14,5))
(df_opl['Equipment'].value_counts() / len(df_opl)).plot.area()
```

#### Гистограмма

```python
# визуализация интервальных данных. сколько значений попадает в интервал 
fig = plt.figure(figsize=(14,5))
df_opl['Age'].plot.hist(bins=10) # 10 интервалов 
```

```python
# зависимость одного показания от другова
# условие: < 50 лет, случайные 300 значений. X и Y - оси
df_opl[df_opl['Age']<50].sample(300).plot.scatter(x='BodyweightKg', y='Age', figsize=(14,5))
```

```python
# !!! плохо для большого количества данных !!!
df_opl[df_opl['Age']<50].plot.scatter(x='BodyweightKg', y='Age', figsize=(14,5))
```

```python
# лучше hexbin - 
# gridsize=25 и cmap='Blues (по умолчанию зеленый цвет)
df_opl[df_opl['Age']<50].plot.hexbin(x='BodyweightKg', y='Age', figsize=(17,6), gridsize=25, cmap='Blues')
```
