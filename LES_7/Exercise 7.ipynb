### 1. Точечная диаграмма
Цель этой задачи-создать простую точечную диаграмму, используя случайные входные значения и случайные цвета! Наконец, вы сохраните сюжет в виде файла .png, который должен выглядеть примерно так:

![](img/problem1_example.png)

**Примечание:** ваши цвета могут отличаться от изображения выше в зависимости от версии Matplotlib, которую вы используете!

#### Решение:

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Текущая директория
ROOT_PATH = os.path.abspath(os.curdir)
```

```python
# Создайте новый фрейм данных под названием data и добавьте 1000 случайных чисел (float) в новый столбец x 
# и еще 1000 случайных чисел (float) в новый столбец y.

data = pd.DataFrame(np.random.rand(1000, 2), columns=['x','y'])
print(data.head())
print(data.info())
```

```python
# проверка количества строк
assert len(data) == 1000, "Должно быть 1000 строк данных." 
```

```python
# Создайте переменные цвета, в которых вы должны хранить 1000 случайных цветов.
# Переменные цвета должны быть массивом, полным случайных чисел.
# RGB
colors = pd.DataFrame(np.random.rand(1000, 3), columns=['R','G', 'B'])
print(colors.head())
print(colors.info())
```

```python
colors.plot()
```

```python
print(colors[0:10])
# проверка количества строк
assert len(colors) == 1000, "Должно быть 1000 строк данных." 
```

#### Создайте точечную диаграмму со случайными цветами
Построить точечную диаграмму, используя таблицу DataFrame.plot():

выберите тип графика, используя параметр kind=scatter (или использование таблицы DataFrame.plot.scatter())
измените размер точки с помощью параметра s (в примере графика используется s=50)
установить случайные цвета для очков с помощью параметра c выбрать цветовую схему, используя цветовую карту параметр.
В примере изображения используется цветовая карта "Радуга"
вы также можете выбрать другую цветовую карту отсюда: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
вы можете добавить черный контур для точек, используя параметр edgecolor. В этом примере используется edgecolor=black.
